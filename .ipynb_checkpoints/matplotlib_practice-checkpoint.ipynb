```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept","Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month. 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## Create the plot MATLAB

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

## Create the plot with ax.plt( ) OO Method

```python
# Create the plot with ax.plt() OO Method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot() OO Method
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() OO Method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

## Create the plot. MATLAB

```python
# Create the plot. MATLAB
plt.plot(x_axis, y_axis)
plt.show()
```

# 5.1.4

## Add labels, legend, markers, title, grid with MATLAB

```python
# Create the plot and add a label for the legend MATLAB.
plt.plot(x_axis, y_axis, label="Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a legend.
plt.legend()
```

```python
# Create the plot and use MATLAB.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add a legend
plt.legend()
```

## Add labels, legend, markers, title, grid with OO Method

```python
ax = plt.axes() OO Method
# 1 line color green, 2 marker diamond, 3 line width is 2pts
ax.plot(x_axis, y_axis, label='line', color='green', marker='D', linewidth=2)

# 4 add legend for Boston
ax.legend(['Boston'])

# 5 add title and axes labels
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')

# 6 add a y-axis limit
ax.set_ylim(0, 45)

# 7 add grid lines
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

## SKILL DRILL: Scatter plot using MATLAB

```python
plt.scatter(y_axis, x_axis, color='red')
plt.legend(["Chicago"])
plt.title("PyBer Fare by Month")
plt.xlabel("Fares ($)")
plt.ylabel("Date")
plt.xlim(0, 45)
plt.gca().invert_yaxis()
```

## Create Bubble Chart using MATLAB

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

### Make all bubbles larger

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)    
```

### Using list comprehension

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Change the color of the markers to sky blue.
fig, ax = plt.subplots()
#ax.scatter(x_axis, y_axis, color='blue')

#Change the size of the markers to 5 times each data point.
#Make the color 20% transparent.
#ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], c='blue', alpha=0.2)

#Add a black edge color to the circles.
#ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], c='blue', alpha=0.2, edgecolor='black')

#Make the linewidth of the circles 2 points.
#ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], c='blue', alpha=0.2, edgecolor='black', linewidths=2)

#Add a title.
ax.set_title("PyBer Fare by Month")

#Switch the axis so that the Fare($) data is on the x-axis.
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], alpha=0.8, c='skyblue', edgecolor='black', linewidth=2)
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")

#Add a legend for the city of Boston.
ax.legend(["Boston"])

#Add a limit for the x-axis data.
ax.set_xlim(0, 50)

#Invert the y-axis so that January is at the top.
ax.invert_yaxis()
```

## Create Pie Charts

## MATLAB Method

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values=[0,0,0,0,0,0,0.2,0,0,0,0,0]
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct='%.1f%%')
plt.show()
```

### Assign colors to each wedge

```python
# Assign 12 colors, one for each wedge
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')
plt.show()
```

## Create pie chart using OO Method & SKILL DRILL

```python
# Create pie chart
fig, ax=plt.subplots()

#Add a percentage to one decimal place to each wedge of the pie.
#Increase the figure size to 8x8.
fig.set_size_inches(8,8)

#Explode the two highest percentage months.
explode_values=(0,0,0.2,0,0,0,0.2,0,0,0,0,0)

#Add new colors of your own choosing or use the colors from the previous pie chart.
colors=['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'coral', 'mediumpurple', 'lightgreen', 'cornflowerblue', 
       'papayawhip', 'tab:cyan', 'orchid', 'bisque']
#Add a shadow.
#Add a start angle so that January is at the top.
#Reverse the order so that the month order is in a clockwise direction.
ax.pie(y_axis,
    labels=x_axis,
    autopct='%.1f%%',
    explode=explode_values,
    shadow=True,
    colors=colors,
    startangle=(95),
    counterclock=False)
plt.show()
```
